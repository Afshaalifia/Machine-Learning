## 1. Tentukan Machine Learning

**1. Time Series Forecasting (Peramalan Deret Waktu)**
- Metode prediksi yang digunakan untuk memproyeksikan nilai-nilai masa depan dari suatu variabel berdasarkan data historis yang tersusun secara berurutan berdasarkan waktu. Data time series adalah kumpulan pengamatan atau data yang dikumpulkan pada interval waktu tertentu, misalnya harian, bulanan, atau tahunan. Contoh umum dari data time series adalah harga saham, permintaan listrik, produksi pabrik, dan suhu cuaca.


##2. Sertakan alasan penggunaan/pemilihan machine learning model pada poin 1

- Data Berurutan Secara Waktu: Jika data yang dimiliki berurutan berdasarkan waktu, dan ada pola yang jelas yang muncul secara teratur seiring waktu (seperti tren atau musiman), maka time series forecasting adalah metode yang paling tepat. Dalam situasi ini, data dari waktu sebelumnya dapat memberikan informasi yang sangat berharga tentang nilai-nilai di masa depan.

- Kebutuhan untuk Prediksi Jangka Pendek atau Menengah: Time series forecasting umumnya sangat efektif untuk prediksi jangka pendek atau menengah. Misalnya, jika Anda perlu memprediksi penjualan untuk bulan depan atau permintaan energi di minggu berikutnya, model time series dapat memberikan hasil prediksi yang cukup akurat karena pola musiman atau tren dapat dikenali dalam jangka waktu ini.

- Efisiensi dalam Mengatasi Masalah Musiman atau Siklus: Jika data memiliki sifat musiman, seperti permintaan produk yang meningkat saat liburan, time series forecasting dapat membantu mengidentifikasi dan memperhitungkan siklus musiman ini.

### Code Machine Learning

```python
# Import pustaka yang diperlukan
import pandas as pd
import warnings
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Menonaktifkan pesan peringatan yang tidak diinginkan
warnings.filterwarnings("ignore", message="Non-stationary starting autoregressive parameters")
warnings.filterwarnings("ignore", message="Non-invertible starting MA parameters")
warnings.filterwarnings("ignore", message="Maximum Likelihood optimization failed to converge")

# Memuat dataset
data = pd.read_csv('cleaned_pivot_green_growth_mys.csv', index_col='time_period', parse_dates=True)

# Mengisi nilai yang hilang menggunakan metode yang diperbarui
data = data.ffill().bfill()

# Memilih measure yang akan diprediksi
target_measure = 'CO2_DBEM (Total - all activities)'
# Mengambil data time series untuk target
y = data[target_measure]

# Memastikan data memiliki frekuensi yang ditentukan
y = y.asfreq('YS')  # Mengasumsikan data tahunan yang dimulai pada bulan Januari

# Menggunakan data asli tanpa differencing
print("Menggunakan data asli tanpa differencing.")

# Memisahkan data menjadi data training dan testing
train_size = int(len(y) * 0.8)
train, test = y[:train_size], y[train_size:]

# Inisialisasi dan melatih model ARIMA dengan data asli
model = ARIMA(train, order=(1, 1, 1))  # Anda bisa mencoba berbagai kombinasi (p, d, q) sesuai hasil ACF dan PACF
model_fit = model.fit()

# Melakukan prediksi menggunakan data testing
predictions = model_fit.forecast(steps=len(test))

# Mengevaluasi kinerja model
mae = mean_absolute_error(test, predictions)
mse = mean_squared_error(test, predictions)
rmse = mse ** 0.5

# Menampilkan hasil evaluasi
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Menampilkan beberapa hasil prediksi pertama beserta data aslinya
results_df = pd.DataFrame({'Actual': test, 'Predicted': predictions}, index=test.index)
print(results_df.head())

# Membuat prediksi masa depan hingga tahun 2030
future_steps = 2030 - data.index[-1].year + 1  # Menghitung jumlah tahun ke depan
future_predictions = model_fit.forecast(steps=future_steps)

# Menghasilkan tanggal masa depan dengan frekuensi yang benar
future_years = pd.date_range(start=data.index[-1] + pd.DateOffset(years=1), periods=future_steps, freq='YS')
future_results = pd.DataFrame({'Year': future_years.year, 'Predicted': future_predictions})

# Menampilkan prediksi hingga tahun 2030
print("\nPrediksi hingga tahun 2030:")
print(future_results)
```

## 4. Lakukan Cross-Validation Basic dari machine learning model yang telah dibuat sebelumnya

```python
# Import pustaka yang diperlukan
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Fungsi untuk melakukan cross-validation
def time_series_cross_validation(data, target_measure, p, d, q, n_splits=5):
    y = data[target_measure]
    y = y.asfreq('YS')  # Mengatur frekuensi data

    # Panjang setiap fold dalam cross-validation
    fold_size = int(len(y) / n_splits)

    # Menyimpan hasil evaluasi
    mae_scores = []
    mse_scores = []
    rmse_scores = []

    for i in range(1, n_splits):
        # Tentukan train dan test set berdasarkan split
        train = y[:fold_size * i]
        test = y[fold_size * i: fold_size * (i + 1)]

        # Melatih model ARIMA pada data train
        model = ARIMA(train, order=(p, d, q))
        model_fit = model.fit()

        # Prediksi pada data test
        predictions = model_fit.forecast(steps=len(test))

        # Evaluasi prediksi dengan MAE, MSE, dan RMSE
        mae = mean_absolute_error(test, predictions)
        mse = mean_squared_error(test, predictions)
        rmse = mse ** 0.5

        # Simpan hasil evaluasi
        mae_scores.append(mae)
        mse_scores.append(mse)
        rmse_scores.append(rmse)

    # Rata-rata hasil evaluasi
    mean_mae = np.mean(mae_scores)
    mean_mse = np.mean(mse_scores)
    mean_rmse = np.mean(rmse_scores)

    # Menampilkan hasil cross-validation secara keseluruhan
    print("\nHasil Cross-Validation:")
    print(f"Rata-rata MAE: {mean_mae}")
    print(f"Rata-rata MSE: {mean_mse}")
    print(f"Rata-rata RMSE: {mean_rmse}")

# Memanggil fungsi cross-validation
time_series_cross_validation(data, target_measure='CO2_DBEM (Total - all activities)', p=1, d=1, q=1, n_splits=5)
```

RMSE adalah akar kuadrat dari MSE dan mengembalikan nilai kesalahan ke dalam satuan yang sama dengan data asli. Nilai RMSE sebesar 35.56 menunjukkan bahwa, secara rata-rata, kesalahan prediksi model ARIMA adalah sekitar 35.56 unit. Seperti halnya MAE, semakin kecil nilai RMSE, semakin baik kinerja model.

**Kesimpulan:**
Nilai MAE dan RMSE yang cukup besar (34.82 dan 35.56) menunjukkan bahwa prediksi model ARIMA ini memiliki tingkat kesalahan yang signifikan. Kesalahan ini dapat dianggap besar, terutama jika data asli memiliki skala yang lebih kecil. Jika data aktual berada dalam rentang nilai yang lebih tinggi, kesalahan ini mungkin dapat diterima. Namun, jika rentang data relatif kecil, maka model perlu diperbaiki atau dioptimalkan lebih lanjut. Upaya perbaikan bisa meliputi penyesuaian parameter model atau mencoba model yang berbeda untuk meningkatkan akurasi prediksi.

## 10. Berikan interpretasi dari hasil metric evaluation serta alasan pemilihan metric yang dipilih pada poin 9

Berikut adalah interpretasi dari hasil metric evaluation pada model yang telah dibuat:

- Accuracy: Mengukur seberapa sering prediksi benar dari total prediksi. Jika accuracy tinggi (misalnya >0.9), maka model mampu memprediksi dengan baik untuk data yang seimbang.
- Precision: Penting jika kita ingin meminimalkan false positives. Jika precision tinggi, berarti sedikit kesalahan dalam mengklasifikasikan kelas positif.
- Recall: Penting jika kita ingin mendeteksi sebanyak mungkin kelas positif (false negatives harus sedikit). Recall tinggi menunjukkan model efektif dalam menangkap semua kasus positif.
- F1-Score: Rata-rata harmonis antara precision dan recall. Dipilih ketika distribusi kelas tidak seimbang, karena memberikan keseimbangan antara false positives dan false negatives.

Alasan Pemilihan Metrik:
- Accuracy baik digunakan jika kelas seimbang, namun tidak cukup jika data memiliki ketidakseimbangan kelas.
- Precision dan Recall memberikan pandangan lebih dalam terhadap kinerja model di kelas minoritas, sehingga lebih sesuai jika ada perbedaan signifikan antara distribusi kelas.
- F1-Score sangat penting untuk data yang memiliki distribusi kelas tidak seimbang, karena memastikan bahwa model baik dalam menemukan kelas minoritas tanpa mengorbankan akurasi keseluruhan.

Kombinasi dari metrik-metrik ini memberikan gambaran lebih lengkap mengenai performa model, terutama pada masalah klasifikasi dengan distribusi data yang tidak merata.
